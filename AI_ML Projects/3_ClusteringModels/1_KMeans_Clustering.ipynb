# K Means Clustering

## Importing libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing dataset

```
dataset = pd.read_csv('AmazonClusteringModel.csv')
X = dataset.iloc[:, 3:5].values #We're only taking the last two columns from the dataset for now, as we are only going to be ploting it in a 2D Graph
```

```
print(X)
```

## Optimal number of clusters via Elbow Method

```
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=21)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title('No of clusters using Elbow Method')
plt.xlabel('No. of clusters')
plt.ylabel('Wcss')
plt.show()
```

## K Means Model Training on Training set

```
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 21)
y_means = kmeans.fit_predict(X)
```

```
print(y_means)
```

## Visualizing Clusters

```
plt.scatter(X[ y_means == 0, 0], X[y_means == 0, 1], s = 100, c = 'red', label = 'Cluster 0') #s==size, c==color
plt.scatter(X[y_means == 1, 0], X[y_means == 1, 1], s = 100, c = 'blue', label = 'Cluster 1')
plt.scatter(X[y_means == 2, 0], X[y_means == 2, 1], s = 100, c = 'green', label = 'Cluster 2')
plt.scatter(X[y_means == 3, 0], X[y_means == 3, 1], s = 100, c = 'yellow', label = 'Cluster 3')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 200, c = 'black', label = 'Centroids') #also ploting the centroids of each clusters

plt.title('Clustering the Ratings based on Income')
plt.xlabel('Income')
plt.ylabel('Rating')
plt.legend()
plt.show()
```
